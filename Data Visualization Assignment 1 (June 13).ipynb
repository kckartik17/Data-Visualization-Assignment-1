### Question 4 - Build a program to predict the correct word for wrongly typed word. for example whst => what, numbr =>number, good => good

```python
from spellchecker import SpellChecker
words = []
with open('big.txt','r') as f:
    words = f.readlines()
words_list = []
for word in words:
    words_list += word.rstrip("\n").split(" ")
spell = SpellChecker()
spell.word_frequency.load_words(words_list)
def correct_spelling(entered_word):
    misspelled = spell.unknown([entered_word])
    if len(misspelled) != 0:
        for word in misspelled:
            print("Correct Spelling: {}".format(spell.correction(word)))
    else:
        print(entered_word)
```

```python
correct_spelling("peterbrg")
```

```python
correct_spelling("numbr")
```

```python
correct_spelling("good")
```

```python
correct_spelling("invesgate")
```
